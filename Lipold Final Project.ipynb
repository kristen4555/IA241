## Background Information on COVID-19

COVID-19, also known as the Coronavirus, is a contagious disease spread through infected droplets and particles. Most individuals who find themselves sick with the virus have mild symptoms. However, those who are elderly or immunocompromised are at more risk for severe effects, including death. The best way to protect yourself from the virus are to get vaccinated, minimize the spread of germs, and avoid close contact with those who are currently contagious. 
Information taken from [Centers for Disease COntrol and Prevenetion](https://www.cdc.gov/coronavirus/2019-ncov/faq.html#Basics).

## Sum of Spanish COVID-19 Cases per Month in 2020

```python
df = pandas.read_excel('s3://ia241-spring2022-lipold/covid_data.xls')
spain_data = df.loc[df['countriesAndTerritories'] == 'Spain'] 
sum_cases_per_month = spain_data.groupby('month').sum()['cases']
```

```python
sum_cases_per_month.plot.bar()
```

This bar graph highlights the spike in Spanish COVID-19 cases following summer 2020 and reaching it's peak in November. The sharp drop in cases can partially be explained by the [European Centre for Disease Prevention and Control](https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide) no longer tracking Coronavirus data following December 14th, 2020.

## Total Number of Cases Confirmed in Spain in 2020

```python
sum_cases = spain_data.sum()['cases']
print(sum_cases)
```

## Number of Confirmed Deaths in Spain in 2020

```python
sum_deaths = spain_data.sum()['deaths']
print(sum_deaths)
```

## How the Number of Cases is Related to Number of Deaths in Spain

```python
spain_data.plot.scatter(x='cases',y='deaths',c='month')
```

Calculating the confirmed trend between the number of deaths and cases allows Spain to be measured against other countires in terms of the effectiveness of safety protocals. The relationship between the confirmed cases and deaths shows Spain's ability care for infected individuals.
